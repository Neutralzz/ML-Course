## 偏差与方差 Bias and Variance Error

为了解决一个问题，我们希望得到生成这个问题的模型 $\hat f$ ，进行 $N$ 次实验，每次实验都从采集数据开始（训练集不同），直到选出最佳的模型 $ f^* $ （$ f^* $ 是 $\hat f$ 的估计），从而有了大小为 $N$ 的模型集合 $ F = \{ f^{*1},f^{*2},f^{*3},...,f^{*N} \} $。

偏差 Bias Error 是 $F$ 内模型的均值 $\bar f$ 与 $ \hat f $的距离：
$$ E[f^*] = \bar f $$
$$ Err_{bias}[f^*] = \bar f - \hat f $$

方差 Variance Error 是$F$ 内各个模型 $f^*$ 与均值 $\bar f$ 的均方距离。
$$ Err_{var}[f^*] = \frac{1}{N} \Sigma_{n=1}^N (f^{*n} - \bar f)^2$$

# 处理误差

* 如果模型不能很好的拟合训练集，那么说明偏差较大。
* 如果模型能很好地拟合训练集，但是在测试集上有很大的误差，说明方差较大。

## 处理偏差
* 在输入中增加更多的特征
* 选择一个更复杂的模型

## 处理方差
* 更多的数据
* 正则化
