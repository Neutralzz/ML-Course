## 3.1 回归 Regression

目标函数 $f$ 的输出是数值类型的 "scalar"。

比如预测明天上午的PM2.5，$f$ 输入是最近的PM2.5，$f$ 输出是明天PM2.5（数值类型）。

## 3.2 分类 Classification

分类问题有二分类和多分类。

二分类就是判断是与不是，比如识别邮件是否为垃圾邮件；

多分类就是从不同选项里做选择，比如识别单个数字，有0到9总共10个选择；

## 3.3 Structured Learning

目标函数的output，不再是简单的数值或者是类别，而是具有结构性质的结果。

比如语音识别，输出是一句话；机器翻译，输出是输入的另一种语音形式等等。

## 3.4 监督学习之外

### 3.4.1 半监督学习 Semi-supervised Learning

以识别猫和狗为例，监督学习的数据集是都有标注是猫还是狗，半监督学习的数据是有一部分是标注的，其余是<strong>未标注的猫和狗的图片</strong>

### 3.4.2 迁移学习 Transfer Learning

区别于半监督学习，其数据是部分标注猫狗图片，但其余是<strong>标注或未标注的可能与任务本身无关的图片</strong>




### 3.4.3 无监督学习 Unsupervised Learning

In machine learning , the problem of unsupervised learning is that of trying to find hidden structure in unlabeled data.  --Wiki

区别于其他Learning，其数据是没有标签的unlabeled，机器学习要直接通过这些数据建模。

比如，通过阅读大量的文章来理解单词的意思；通过观察大量的图片来画某个物品。

###  3.4.4 强化学习 Reinforcement Learning

强化学习是在尝试的过程中学习到在特定的情境下选择哪种行动可以得到最大的回报。

强化学习过程中，机器一开始是什么也不知道的，通过学习来知道自己的行为是好还是不好。

它采用的是边获得样例边学习的方式，在获得样例之后更新自己的模型，利用当前的模型来指导下一步的行动，下一步的行动获得Reward之后再更新模型，不断迭代重复直到模型收敛。

以聊天机器人为例区分监督学习和强化学习：

   监督学习下，当人对机器说"Hello"，机器知道要回复"Hi"，而强化学习下，机器因为不知道回复什么，就在随便交流，最后把人气跑了，从而得知自己的行为是不对的，再更新模型。
